```python
# -*- coding: utf-8 -*-
"""GAN_3.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1uNbxVaZrMu5F0XNhRmCIlZxX66T1zm7Y
"""

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Activation, BatchNormalization, Dense, Dropout, Flatten, Reshape
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import Conv2D, Conv2DTranspose
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

import warnings , os

os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

#Implementacin de Deep Convolutional
#representaremos tanto al Generador como al Discriminador como redes convolucionales
(x_train, y_train), (_, _) = mnist.load_data()
print(type(x_train))

#Dimensiones de entrada del modelo
img_rows = 28
img_cols = 28
channels = 1

# Input image dimensions
img_shape = (img_rows, img_cols, channels)
z_dim = 100                                     #Tamao del vector de ruido, utilizado como entrada al Generador

#1.-Tome un vector de ruido aleatorio y remodelelo en un tensor de 7  7  256
#2.-Utilice la convolucin transpuesta, transformando el tensor de 7  7  256 en un tensor de 14  14  128.
#3.-Aplique la normalizacin por lotes y la funcin de activacin de Leaky ReLU.
#4.-Utilice la convolucin transpuesta, transformando el tensor de 14  14  128 en un tensor de 14  14  64. Observe4
#5.-Aplique la normalizacin por lotes y la funcin de activacin de Leaky ReLU.
#6Use la convolucin transpuesta, transformando el tensor de 14  14  64 en el tamao de la imagen de salida, 28  28  1.
#7Aplique la funcin de activacin de tanh.

#strides- especificando los pasos de la convolucin a lo largo de la altura y el ancho. Puede ser un entero nico para especificar el mismo valor para todas las dimensiones espaciales

#DCGAN Generator
def build_generator(z_dim):

    model = Sequential() #modelo es apropiado para una simple pila de capas donde cada capa tiene exactamente un tensor de entrada y un tensor de salida
    model.add(Dense(256 * 7 * 7, input_dim=z_dim))
    model.add(Reshape((7, 7, 256)))
    model.add(Conv2DTranspose(128, kernel_size=3, strides=2, padding='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.01))
    model.add(Conv2DTranspose(64, kernel_size=3, strides=1, padding='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.01))
    model.add(Conv2DTranspose(1, kernel_size=3, strides=2, padding='same'))
    model.add(Activation('tanh'))

    return model

#1.- Use una capa convolucional para transformar una imagen de entrada de 28  28  1 en un tensor de 14  14  32.
#2.- Aplique la funcin de activacin de Leaky ReLU. 3 Utilice una capa convolucional, transformando el tensor de 14  14  32 en un tensor de 7  7  64.
#3.- Aplique la normalizacin por lotes y la funcin de activacin de Leaky ReLU
#4.- Utilice una capa convolucional, transformando el tensor de 7  7  64 en un tensor de 3  3  128.
#5.- Aplique la normalizacin por lotes y la funcin de activacin de Leaky ReLU.
#6.- Aplane el tensor 3  3  128 en un vector de tamao 3  3  128 = 1152
#7 Utilice una capa completamente conectada que alimente la funcin de activacin sigmoidea para calcular la probabilidad de que la imagen de entrada sea real.

#DCGAN Discriminator
def build_discriminator(img_shape):
    model = Sequential()
    model.add(Conv2D(32,kernel_size=3,strides=2,input_shape=img_shape,padding='same'))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Conv2D(64,kernel_size=3,strides=2,padding='same'))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Conv2D(128,kernel_size=3,strides=2,padding='same'))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))
    return model

def build_gan(generator, discriminator):
    model = Sequential()

    # Genearator -> Discriminator
    model.add(generator)
    model.add(discriminator)

    return model


# Creacin y compilacin de modelos discriminadores
discriminator = build_discriminator(img_shape)
discriminator.compile(loss='binary_crossentropy',optimizer=Adam(), metrics=['accuracy'])
# Preserve los parmetros del discriminador mientras entrena al generador
discriminator.trainable = False

# Crear un modelo generador
generator = build_generator(z_dim)

# Cree y compile un modelo GAN con discriminadores congelados para entrenar al generador
gan = build_gan(generator, discriminator)
gan.compile(loss='binary_crossentropy', optimizer=Adam())

from operator import ge
losses = []
accuracies = []
iteration_checkpoints = []

def train(iterations, batch_size, sample_interval):
    # Cargar conjunto de datos MNIST
    (x_train, y_train), (_, _) = mnist.load_data()

    # [0, 255] escala los valores de pxeles en blanco y negro entre [-1, 1]
    X_train = x_train / 127.5 - 1.0
    X_train = np.expand_dims(X_train, axis=3)

    # Etiqueta de imagen real: All 1
    real = np.ones((batch_size, 1))

    # Etiquetas de imgenes falsas: Todas 0
    fake = np.zeros((batch_size, 1))

    for iteration in range(iterations):
        # -------------------------
        #  Entrenamiento discriminatorio
        # -------------------------

        # Obtenga lotes aleatorios de imgenes reales
        idx = np.random.randint(0, X_train.shape[0], batch_size)

        imgs = X_train[idx]

        # Crear lotes de imgenes falsas
        z = np.random.normal(0, 1, (batch_size, 100))
        gen_imgs = generator.predict(z)


        # Training
        d_loss_real = discriminator.train_on_batch(imgs, real)
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
        d_loss, accuracy = 0.5 * np.add(d_loss_real, d_loss_fake)

        # ---------------------
        #  Training generator
        # ---------------------

        # Crear lotes de imgenes falsas
        z = np.random.normal(0, 1, (batch_size, 100)) #z class 'numpy.ndarray'
        gen_imgs = generator.predict(z)

        if iteration == 0:
            print("\nRandom noise input image")
            sample_images(generator)

        g_loss = gan.train_on_batch(z, real)

        if iteration < 200 and (iteration + 1) % 20 == 0:
            # Ahorre prdida y precisin para trazar grficos despus del entrenamiento
            losses.append((d_loss, g_loss))
            accuracies.append(100.0 * accuracy)
            iteration_checkpoints.append(iteration + 1)

            print("%d [D loss: %f, accuracy: %.2f%%] [G loss: %f]" %
                  (iteration + 1, d_loss, 100.0 * accuracy, g_loss))

            # Salida de muestra de imagen generada
            sample_images(generator)

        elif (iteration + 1) % sample_interval == 0:
            #Ahorre prdida y precisin para trazar grficos despus del entrenamiento
            losses.append((d_loss, g_loss))
            accuracies.append(100.0 * accuracy)
            iteration_checkpoints.append(iteration + 1)

            print("%d [D loss: %f, accuracy: %.2f%%] [G loss: %f]" %
                  (iteration + 1, d_loss, 100.0 * accuracy, g_loss))

            # Salida de muestra de imagen generada
            sample_images(generator)

def sample_images(generator, image_grid_rows=5, image_grid_columns=5):

    # muestreo de ruido aleatorio
    z = np.random.normal(0, 1, (image_grid_rows * image_grid_columns, z_dim))

    # Generar imagen a partir de ruido aleatorio
    gen_imgs = generator.predict(z)

    # Escala los valores de pxeles de la imagen entre [0, 1]
    gen_imgs = 0.5 * gen_imgs + 0.5

    plt.figure(figsize=(6,4))
    sns.distplot(gen_imgs,color='blue',kde=True)
    plt.title("Generated Images Distribution")
    plt.show()

    fig, axs = plt.subplots(image_grid_rows,
                            image_grid_columns,
                            figsize=(7, 7),
                            sharey=True,
                            sharex=True)

    cnt = 0
    plt.figure(figsize=(9,9))
    for i in range(image_grid_rows):
        for j in range(image_grid_columns):
            axs[i, j].imshow(gen_imgs[cnt, :, :, 0], cmap='gray')
            axs[i, j].axis('off')
            cnt += 1
    plt.show()
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#    axs[1, 1].imshow(gen_imgs[2, :, :, 0], cmap='gray')
#    axs[1, 1].axis('off')
#plt.show()

#Listing 4.8 Running the model
iterations = 20000
batch_size = 128
sample_interval = 1000 #Entrena el DCGAN para el nmero especificado de iteraciones
train(iterations, batch_size, sample_interval)

n=1
n_z=100
z_sample = np.random.randn(n,n_z)
x_gen = generator.predict(z_sample)

plt.figure(figsize=(5,20))

for i in range(1,n+1):
  plt.subplot(1,n,i)

  plt.imshow(x_gen[i-1].reshape(28, 28), cmap='Greys_r')
plt.show()
```
