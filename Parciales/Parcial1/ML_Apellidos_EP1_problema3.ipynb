```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
dataset = pd.read_csv('dataset2.csv', header=None)
```

```python
D = dataset.iloc[:, :4].values

# Medias
mean = np.mean(D, axis=0)

#Matriz de covarianza de la matriz D
C = np.cov(D, rowvar=False, bias=True)
print(C)

# eigenvalores y eigenvectores
eigval, eigvec = np.linalg.eig(C)
scale_factor = 0.8

# Gráfica
plt.figure()
plt.scatter(D[:, 0], D[:, 1])

# Gráfica eigenvectores
for i in range(len(eigval)):
    plt.arrow(mean[0], mean[1], eigvec[0, i] * np.sqrt(eigval[i]) * scale_factor,
              eigvec[1, i] * np.sqrt(eigval[i]) * scale_factor,
              width=0.005, color="C" + str(i), lw=1, overhang=0.1)

plt.grid(True)
plt.axis('equal')
plt.show()
```
